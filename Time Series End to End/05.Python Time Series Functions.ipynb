# Python Time Series Useful Functions

```python
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime
import requests

mat.rcParams['figure.figsize'] = (10,8)
mat.rcParams['axes.grid'] = False
```

```python
coindeskURL = "https://api.coindesk.com/v1/bpi/historical/close.json?"
```

```python
start = datetime.date(2017, 1, 1)
end = datetime.date(2020, 7, 2)

url = f'{coindeskURL} start={start:%Y-%m-%d}&end={end:%Y-%m-%d}'
result = requests.get(url)
result.content
```

```python
data = pd.read_json(result.content).iloc[:-2, :1]
data.index.name = 'date'
data.index = pd.to_datetime(data.index) # CONVERTING INTO DATE TIME COLUMN
data # WE HAVE DATA FROM 2017-2020 
```

```python
data.shape # ONLY THE bpi COLUMN IS THERE
```

```python
data.plot() # JUST PLOTTING THE TIME SERIES DATA
```

### Grouping the data at different levels then taking the mean of it

```python
data.groupby(data.index.dayofweek).mean() # GROUPING BY DAYS OF WEEK THEN TAKING THE MEAN OF IT
```

```python
data.groupby(data.index.month).mean() # GROUPING BY MONTH LEVEL THEN TAKING THE MEAN OF IT
```

```python
data.groupby(data.index.year).mean() # GROUPING BY YEAR LEVEL THEN TAKING THE MEAN OF IT
```

### using cumsum()

     It will just give you the sum upto the previous values

```python
data[:5]
```

```python
data[:5].cumsum() 
```

### pct_change()

```python
data[:5]
```

```python
data[:5].pct_change(2)
```

### expanding()

    It is very similar to  cumsum but here you can perform mean or any other agg functions as well

```python
pd.concat([data[:5].expanding().sum(), data[:5].cumsum()],axis=1)
```

```python
data[:5].expanding().sum()
```

### Resampling Quarterly then taking the mean of it

```python
data.resample('Q').mean()
```

```python
data.resample('Q').mean().plot() # PLOTTING THE SAME ABOVE DATA
```

```python
data.plot() # NORMAL DATA PLOT
data.resample('M').mean().plot() # RESAMPLING MONTH WISE THEN PLOTTING 
```

```python
data[:5] # TAKING ONLY FIRST 5 ROWS
```

**IT WILL TAKE THE DATA AND SUBTRACT IT TO THE PREVIOUS DATA**

```python
data[:5].diff() # IT WILL TAKE THE DATA AND SUBTRACT IT TO THE PREVIOUS DATA
```

**HERE IT WILL DIFFERENCE THE PREVIOUS 2**

```python
data[:5].diff(2) # HERE IT WILL DIFFERENCE THE PREVIOUS 2 
```

```python
# PLOTTING ALL THE DIFFERENCE ... WE CAN SEE THAT WE HAVE CONVERTED THESE A KIND OF STATIONARY DATA

pd.concat([data['bpi'], data['bpi'].diff(2), data['bpi'].diff()], axis=1).plot()
```

```python
data[:5].shift() # IT JUST SHIFTS THE INDEX ... WITH THE DESIRED VALUE MENTIONED IN THAT
```

```python
data[:5].shift(2) # SHIFTING BY 2 INDEXES
```

```python
data[:5].tshift(2)
```

**SAME THING THE diff() FUNCTION DOES**

```python
data[:5] - data[:5].shift(2)
```

```python
pd.concat([data, data-data.shift(), data.tshift(365)], axis=1).plot() # PLOTTING ALL 3 DATA
```

```python
data.rolling(window=10).mean().plot() # WHEN WE WANT TO CALCULATE MOVING AVERAGE
```

#### RESAMPLING WEEK WISE THEN APPLYING THE MOVING AVERAGE
    HERE WE TOOK WEEKLY WISE SO HERE DATA IS LITTLE SMOOTHENED

```python
# HERE WE TOOK WEEKLY WISE SO HERE DATA IS LITTLE SMOOTHENED
data.resample('W').mean().rolling(window=10).mean().plot() # RESAMPLING WEEK WISE THEN APPLYING THE MOVING AVERAGE 
```

```python
data.rolling(window=50).mean().plot() # APPLYING THE MOVING AVERAGE 
data.rolling(window=50, win_type='gaussian').sum(std=10).plot()
```

```python
data.ewm(span=50).mean().plot()
```

```python
pd.concat([data, data.rolling(window=100).mean(), data.ewm(span=100).mean()], axis=1).plot()
```
